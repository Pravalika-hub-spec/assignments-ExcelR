```python
import pandas as pd
```

```python
data=pd.read_csv("C:\\Users\\DELL\\Downloads\\DATASCIENCE ASSIGNMENTS\\Online retail.csv")
```

```python
print(data.columns)
```

```python
from mlxtend.preprocessing import TransactionEncoder
# Convert the column with item lists into a list of lists
transactions = data['shrimp,almonds,avocado,vegetables mix,green grapes,whole weat flour,yams,cottage cheese,energy drink,tomato juice,low fat yogurt,green tea,honey,salad,mineral water,salmon,antioxydant juice,frozen smoothie,spinach,olive oil'].apply(lambda x: x.split(','))
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)
df.head()
```

# Data Preprocessing:

```python
df.isnull().sum().sum()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

# Association Rule Mining:

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.7)
rules
```

```python
rules_sorted = rules.sort_values(by='lift', ascending=False)
rules_sorted
```

# Interview Questions:

# 1.	What is lift and why is it important in Association rules?

The lift value is an important measure in association rule mining because it provides a way to distinguish between significant and insignificant association rules. 
A high lift value indicates that the association rule is more significant, as it suggests that the two items are highly dependent on each other.

# 3.	What are some limitations or challenges of Association rules mining?
In data mining, association rules are useful for analyzing and predicting customer behavior. The primary disadvantages of association rule algorithms are obtaining boring rules, having a large number of discovered rules, and low algorithm performance.
