# 3. Feature Engineering:

```python
enc_df=pd.get_dummies(df,dtype=int)
enc_df.head()
```

# 4. Decision Tree Classification:

```python
from sklearn.model_selection import train_test_split
```

```python
x = enc_df.drop('age_num', axis=1) 
y = enc_df['age_num']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=42)
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report
```

```python
preds_test = model.predict(x_test)
```

```python
print(classification_report(preds_test,y_test))
```

```python
preds_train=model.predict(x_train)
print(classification_report(preds_train,y_train))
```

# 5. Hyperparameter Tuning:

```python
model_ent= DecisionTreeClassifier(criterion = 'entropy')
model_ent.fit(x_train,y_train)
```

```python
model_max_dep = DecisionTreeClassifier(max_depth=3)
model_max_dep.fit(x_train,y_train)
```

```python
model_min_samp= DecisionTreeClassifier(min_samples_split=6)
model_min_samp.fit(x_train,y_train)
```

```python
model_gini = DecisionTreeClassifier(criterion='gini')
model_gini.fit(x_train,y_train)
```

# 6. Model Evaluation and Analysis:

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f'Test Accuracy: {accuracy:.2f}')
print('Classification Report:')
print(report)
```

# Interview Questions:

# 1. What are some common hyperparameters of decision tree models, and how do they affect the model's performance?

The hyperparameters of a decision tree include the maximum size of the tree, the minimum number of instances required in a node for inducing a split, the node splitting criterion, and the amount of pruning . These hyperparameters play a crucial role in determining the performance of the decision tree model.

# 2. What is the difference between the Label encoding and One-hot encoding?

Label encoding assigns a unique numerical value to each category, while one-hot encoding creates binary columns for each category, with only one column being 1 and the rest 0 for each observation.
