# 3.Data Visualization:

```python
#HISTOGRAMS
data.hist(figsize=(14, 10), bins=30, edgecolor='black')
plt.suptitle('Histograms of Numerical Features')
plt.show()
```

```python
#BAR CHART
plt.figure(figsize=(20, 6))
data.plot(kind='bar', color=['blue', 'orange', 'green'])
plt.title('Distribution of Cardiotocographic Classes')
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.xticks(ticks=[0, 1, 2], labels=['Normal', 'Suspect', 'Pathologic'], rotation=0)
plt.show()
```

```python
#SCATTER PLOT
plt.figure(figsize=(10, 6))
plt.scatter(df['ASTV'], df['MSTV'], alpha=0.5)
plt.title('Scatter Plot of ASTV vs MSTV')
plt.xlabel('ASTV ')
plt.ylabel('MSTV')
plt.show()


```

```python
#HEAT MAP
corr = data.corr()
# Plot the heatmap
plt.figure(figsize=(16, 10))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, square=True)
plt.title('Correlation Heatmap of CTG Dataset')
plt.show()
```

```python
#PAIR PLOT
sns.pairplot(data)
```
