# 1. Data Exploration and Preprocessing:

```python
adults_df=adults.dropna()
```

```python
numeric_cols = adults_df.select_dtypes(include=['number']).columns
numeric_cols
```

```python
from sklearn.preprocessing import StandardScaler
scaler_=StandardScaler()
adults_df[numeric_cols] = scaler_.fit_transform(adults_df[numeric_cols])
adults_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
adults_df[numeric_cols] = scaler.fit_transform(adults_df[numeric_cols])
adults_df.head()
```

# 2. Encoding Techniques:

```python
categorical_cols = adults_df.select_dtypes(include=['object']).columns
categorical_cols
```

```python
from sklearn.preprocessing import LabelEncoder
encoded_df = adults_df.copy()
for column in categorical_cols:
    unique_values = encoded_df[column].nunique()
    if unique_values < 5:
        # Apply One-Hot Encoding
        encoded_df = pd.get_dummies(encoded_df, columns=[column], drop_first=True,dtype=int)
    else:
        # Apply Label Encoding
        le = LabelEncoder()
        encoded_df[column] = le.fit_transform(encoded_df[column])
encoded_df
```

```python
One-Hot Encoding

Pros:

No Ordinal Relationship: Suitable for nominal data where there is no natural order between categories.
Model Compatibility: Works well with algorithms that require input data without ordinal relationships (e.g., linear regression, KNN).

Cons:
High Dimensionality: Increases the number of features significantly, especially with variables that have many categories.
Memory and Computational Cost: More memory and computation are required due to the increased number of columns.

    
Label Encoding

    Pros:

Simplicity and Efficiency: Converts categories to numerical values directly, which is memory efficient.
No Increase in Dimensionality: Keeps the number of features unchanged.

    Cons:
Implied Ordinality: Imposes an ordinal relationship between categories, which may not be appropriate for nominal data.
```

# 3. Feature Engineering:

```python
# Feature 1: Interaction between 'education_num' and 'hours_per_week'
adults_df['education_hours_interaction'] = adults_df['education_num'] * adults_df['hours_per_week']

# Feature 2: Age Grouping
adults_df['age_group'] = pd.cut(adults_df['age'], bins=[0, 25, 45, 65, 100], labels=['Young Adult', 'Middle Aged', 'Senior', 'Elderly'])
adults_df.head()
```

```python
Feature 1: 
Rationale: This feature could capture the combined effect of education level and working hours on income. Higher education might yield better returns for those working more hours, and this interaction term could help the model capture this relationship.

Feature 2:
Rationale: Grouping age into categories like "Young Adult", "Middle Aged", and "Senior" could help the model distinguish between different age-related trends. This might capture different income distributions or work habits across age groups.
```

```python
# Check for skewness
print(adults_df[['capital_gain', 'capital_loss']].skew())
```

```python
# Apply log transformation to 'capital_gain'
adults_df['log_capital_gain'] = np.log1p(adults_df['capital_gain'])
adults_df.head()
```

# 4. Feature Selection:

```python
from sklearn.ensemble import IsolationForest
numerical_cols = adults_df.select_dtypes(include=['int64', 'float64']).columns
# Initialize the Isolation Forest model
iso_forest = IsolationForest(contamination=0.05, random_state=42)  
adults_df['outlier'] = iso_forest.fit_predict(adults_df[numerical_cols])
outliers = adults_df[adults_df['outlier'] == 1]
print(adults_df.outlier)
```

```python
# Drop the outliers 
df_cleaned = adults_df[adults_df['outlier'] == 1].drop(columns=['outlier'])
print(f'Number of outliers removed: {len(outliers)}')
```
