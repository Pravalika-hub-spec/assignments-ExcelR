# Task 2: Dimensionality Reduction with PCA:

```python
from sklearn.preprocessing import scale 
```

```python
#normalizing
wine_normal=scale(x)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA()
pca_values=pca.fit_transform(wine_normal)
```

```python
pca_values
```

```python
var=pca.explained_variance_ratio_
var
```

```python
#amout of variance each pca having
variance=pca.explained_variance_ratio_*100
variance
```

```python
# Plot a scree plot
explained_variance_ratio = pca.explained_variance_ratio_
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(feature_names) + 1), explained_variance_ratio, 'o-', markersize=8, color='b')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.show()
```

```python
# Cumulative variance 
cum_var = np.cumsum(np.round(variance,decimals = 4)*100)
cum_var
```

```python
# Cumulative explained variance plot
cumulative_explained_variance = np.cumsum(explained_variance_ratio)
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(feature_names) + 1), cumulative_explained_variance, 'o-', markersize=8, color='green')
plt.xlabel('Principal Component')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance Plot')
plt.show()
```

```python
variences=pd.DataFrame()
variences["percentage"]=pd.Series(pca.explained_variance_ratio_*100)
variences['cumsum']=pd.Series(np.cumsum(np.round(variance,decimals = 4)*100))
variences
```

# Task 3: Clustering with Original Data:

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
#Build Cluster algorithm
clusters_entire = KMeans(n_clusters=3, random_state=42)
kmeans = clusters_entire.fit_predict(wine_normal)
```

```python
# Visualize the clustering result using the first two features
plt.figure(figsize=(10, 6))
plt.scatter(wine_normal[:, 0], wine_normal[:, 1], c=kmeans, cmap='viridis', marker='o')
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])
plt.title('K-means Clustering of Wine Dataset (First Two Features)')
plt.show()
```

```python
# Calculate and print the silhouette score
silhouette_avg = silhouette_score(wine_normal, kmeans)
print(f'Silhouette Score for K-means: {silhouette_avg}')
```

# Task 4: Clustering with PCA Data:

```python
cluster_pca=KMeans(n_clusters=3,random_state=42)
kmeans_pca = cluster_pca.fit_predict(pca_values)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(pca_values[:, 0], pca_values[:, 1], c=kmeans_pca, cmap='viridis', marker='o')
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])
plt.title('K-means Clustering of pca(First Two Features)')
plt.show()
```

```python
silhouette_pca= silhouette_score(pca_values, kmeans_pca)
print(f'Silhouette Score for K-means: {silhouette_pca}')
```

# Task 5: Comparison and Analysis:

```python
1) The original dataset scatter plot uses the first two features, which may not capture the most variance.The PCA-transformed scatter plot uses the first two principal components, which capture the most varianc

2) Both methods aim to cluster data points into meaningful groups. Both visualizations can show clear clusters if the data structure is strong.

3) PCA reduces dimensionality, retaining the most important features.It can improve clustering performance by removing noise and redundant features.
```

# Task 6: Conclusion and Insights

```python
1)Applying PCA before clustering can significantly improve clustering performance by reducing noise and redundancy in the dataset.

2) PCA simplifies the dataset by reducing the number of features while retaining the most important information. This is especially useful for high-dimensional datasets where many features may be redundant or noisy.

3)Use PCA when dealing with high-dimensional data or when clustering performance needs enhancement. 
```
